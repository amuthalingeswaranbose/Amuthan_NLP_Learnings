# Import

```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-miniLM-L6-v2')
model
```

# Embedding

```python
 sentences = [
     "This framework generates embedding for each input sentence",
     "Sentence are passed as a list of stings."
 ]

embeddings = model.encode(sentences)


for sentence, embedding in zip(sentences, embeddings):
  print(f"Sentence: {sentence}")
  print(f"Sentence count: {len(sentence)}, Embedding count: {len(embedding)}")
  print(f"Embedding: {embedding}")
  print(" ")
```

# Cosing - similarity

```python
emb1 = model.encode('I am eating Apple')
emb2 = model.encode('I like fruites')

cos_sim = util.cos_sim(emb1, emb2)

print(f'Cosine Similarity: {cos_sim}')
```
