# Import

```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-miniLM-L6-v2')
model
```

# 1.Embedding

```python
sentences = [
    "This framework generates embedding for each input sentence",
    "Sentence are passed as a list of stings."
 ]

embeddings = model.encode(sentences)


for sentence, embedding in zip(sentences, embeddings):
  print(f"Sentence: {sentence}")
  print(f"Sentence count: {len(sentence)}, Embedding count: {len(embedding)}")
  print(f"Embedding: {embedding}")
  print(" ")
```

# 2.Sentence Similarity (Cosing - similarity)

```python
emb1 = model.encode('I am eating Apple')
emb2 = model.encode('I like fruites')

cos_sim = util.cos_sim(emb1, emb2)

print(f'Cosine Similarity: {cos_sim}')
```

## Compute cosine similarity between paris

```python
sentences = [
    "A man is eating a food.",
    "A man is eating a piece of bread.",
    "The girl is carring a baby.",
    "A man is ridding a horse.",
    "Two men pushed carts through the woods.",
    "A man is riding a horse on a enclosed ground.",
    "A monkey is playing drums.",
    "Someone in a gorilla costume playing a set of drums"
]

embeddings = model.encode(sentences)
print(f"Embedding count: {len(embeddings)}")

cos_siml = util.cos_sim(embeddings, embeddings)

print(f'Cosine Similarity: {cos_siml}')
```

## Add all pairs to a list  with their  consine similarity score

```python
all_sentence_similairity = []

for i in range(0, len(cos_siml)-1):
  for j in range(i+1, len(cos_siml)):
    all_sentence_similairity.append((cos_siml[i][j], i, j))


for sent_sim in all_sentence_similairity:
  print(f"sentence_similairity: {sent_sim}")
```

## Sort list by the highest  consine similarity score

```python
all_sentence_similairity = sorted(all_sentence_similairity, key=lambda x: x[0], reverse=True)
all_sentence_similairity

for score, i, j in all_sentence_similairity[:5]:
  print("{}\t{}\t{:.4f}".format(sentences[i], sentences[j], cos_siml[i][j]))
```

# 4.Clustering

```python
from sklearn.cluster import KMeans
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-V2')
model
```

```python
# Embedd and normalize the corpus

# Corpus with example sentences
corpus = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'Horse is eating grass.',
          'A man is eating pasta.',
          'A Woman is eating Biryani.',
          'The girl is carrying a baby.',
          'The baby is carried by the woman',
          'A man is riding a horse.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.',
          'A cheetah is running behind its prey.',
          'A cheetah chases prey on across a field.',
          'The cheetah is chasing a man who is riding the horse.',
          'man and women with their baby are watching cheetah in zoo'
          ]

# embedding corupus
corpus_embedding = model.encode(corpus)
corpus_embedding
```

# Normalize the embedding to units lenght

```python
corpus_embedding = corpus_embedding / np.linalg.norm(corpus_embedding, axis=1, keepdims=True)
corpus_embedding
```

```python
corpus_embedding[0]
```

```python
clustering_model = KMeans(n_clusters=4)
clustering_model.fit(corpus_embedding)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python
clustered_sentences = {}

for sentence_id, cluster_id in enumerate(cluster_assignment):
  if cluster_id not in clustered_sentences:
    clustered_sentences[cluster_id] = []

  clustered_sentences[cluster_id].append(corpus[sentence_id])

print(f"clustered_sentences: {clustered_sentences}")
```
